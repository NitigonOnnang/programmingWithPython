## **1. ข้อความและการคำนวณ**


> คอมพิวเตอร์เองวัตถุประสงค์เดิมที่ถูกคิดค้นขึ้นมาก็คือเพื่อใช้ช่วยในการคำนวณ และจะเห็นว่ามันสามารถคำนวณได้ในพริบตา ดังนั้นนี่คือประโยชน์อย่างแรกของคอมพิวเตอร์ที่เราจะเห็นได้





>เพียงแต่ว่าในการใช้งานจริงเวลาที่แสดงผลออกมาให้ดูหากมีแค่ตัวเลขอย่างเดียวก็จะดูเข้าใจยาก จึงต้องมีการทำให้แสดงผลตัวหนังสือออกมาด้วย



```python
print('นายภาคภูมิมีเงิน 18000 บาท นำไปเล่นหุ้น 15000 บาท')
print("แต่ขาดทุน ขายคืนมาได้แค่ 3000 บาท นายภาคภูมิเหลือเงิน :")
print("18000 - 15000 + 3000")
print("=")
print(18000 - 15000 + 3000)
```

จากโปรแกรมด้านบนจะเห็นว่า


1.   ข้อความในส่วนที่มีเครื่องหมายคำพูดแบบเดี่ยว `' '` หรือแบบคู่ `" "` ครอบอยู่ จะ`แสดงผลเป็นข้อความ`มาตามที่พิมพ์ไป แม้ว่าจะเป็นตัวเลขก็ตาม
2.   ส่วนที่`ไม่ได้ครอบ`จะ`คำนวณออกมาเป็นตัวเลข`



จากโค้ดด้านล่างจงตอบคำถามต่อไปนี้

1.   นักเรียนคิดว่าเมื่อรันแล้วจะได้ผลเช่นไร ทำไมถึงเป็นเช่นนั้น


```python
text = 'ป้อนคำตอบข้อ1' # @param {type:"string"}
```

2. หากต้องการให้โค้ดทำงานได้ปกติต้องแก้ไขอย่างไร

```python
print(นายภาคภูมิขาดทุนย่อยยับ)
```

## **2. การประมวลผลโค้ด และการคอมเมนต์**

<img src='https://drive.google.com/uc?id=1RH_QSvCqs6l88xO9EGWdgVoHiVrVw8tG'>



> โค้ดที่เขียนลงไปจะถูกอ่านทั้งหมด แต่หากเจอเครื่องหมาย`ชาฟต์ #` โค้ดตรงหลังเครื่องหมายนี้จะกลายเป็น `คอมเมนต์ (comment) `ซึ่งหมายถึงข้อความเขียนประกอบไว้เฉยๆเพื่อเป็นคำอธิบาย ข้อความที่เป็นคอมเมนต์มักจะมีสีต่างไปจากเดิม โดยจะกลายเป็นสีเดียวทั้งหมดไป






```python
print('บ้านกับที่ทำงานอยู่ห่างกัน 120 กิโลเมตร') # s
print('รถวิ่งเร็ว 60 กิโลเมตรต่อชั่วโมง') # v
print('ระยะเวลาที่รถต้องวิ่งคือ') # t
print(120/60) # s=vt ซึ่งก็คือ t = s/v
```

แต่บางครั้งคอมเมนต์ก็ถูกใช้เพื่อเก็บโค้ดที่ไม่ได้ใช้งานแล้วแต่ยังเก็บไว้ก่อนเผื่อใช้อ้างอิง หรืออาจนำกลับมาใช้อีกด้วย เช่น

```python
print('บ้านกับที่ทำงานอยู่ห่างกัน 120 กิโลเมตร') # s
#print('รถวิ่งเร็ว 60 กิโลเมตรต่อชั่วโมง') # v
print('รถวิ่งเร็ว 10 กิโลเมตรต่อชั่วโมง') # รถติดตรงสุขุมวิท
print('ระยะเวลาที่รถต้องวิ่งคือ')
#print(120/60) # s=vt ซึ่งก็คือ t = s/v
print(120/10)
```

จากโค้ดด้านบนหลังจาก run แล้วข้อความแสดงทุกบรรทัดหรือไม่ เพราะเหตุใด
